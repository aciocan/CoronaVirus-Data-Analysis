# Creating a bar chart race animation

With 177 missing rows it's not a good idea to keep the Province/State column in the data.

Also latitude and longitude will not be necessary for creating a bar chart race animation.

```python
chart_df = df.drop(['Province/State','Lat', 'Long'],axis=1)
chart_df.head()
```

```python
chart_df.shape
```

Let's rename the Country/Region column as Country for our future use

```python
chart_df.rename(columns = {'Country/Region':'Country'},inplace=True)
```

```python
chart_df.columns
```

Let's check the number of countries affected by Covid 19

```python
chart_df['Country'].value_counts()
```

So the number of affected countries seem to be 180 but there are 256 columns. So it is very correct to assume that some countries are repeated.

Well thinking more it will be clear that the repetition maybe due to the Provine/State values because more than one provine may belong to the same country

Since all the province belong to the same country it will be safe to get the sum of number of confirmed cases for all the repititive values in Country name.

```python
chart_df.head(10)
```

```python
chart_df = chart_df.groupby('Country').agg(np.sum)
chart_df = chart_df.reset_index().rename(columns = {'index':'Country'})
chart_df.head(10)
```

```python
chart_df.shape
```

Now the data looks ready for creating a bar chart race

```python
#creating the csv file

chart_df.to_csv('bar_chart_race.csv')
```
